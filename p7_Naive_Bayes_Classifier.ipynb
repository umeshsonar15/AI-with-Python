```python
'''
sklearn library contains so many modules and built in ready made features.
load_iris dataset is already preprocessed and transformed into standard form and ready
no need to use pandas or numpy calculations
or standardscaler to transform data
we are all set to go directly to our practical.
Naive Bayes classifier uses conditional probability
P(A|B) = ( P(B|A) x P(A) ) / P(B)
Posterior = ( Prior x Likelihood ) / Evidence
'''
```

```python
from sklearn.datasets import load_iris
X,y = load_iris(return_X_y=True) # splitting input and output
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=0)
# splitting training and testing data into 70% for training and 30% for testing
```

```python
from sklearn.naive_bayes import GaussianNB # for continous values in gaussian distribut
 # i.e normal distribution
clf = GaussianNB()
clf.fit(xtrain,ytrain)
```

```python
ypred = clf.predict(xtest)
ypred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(ytest,ypred)
ac = ac*100
ac
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
