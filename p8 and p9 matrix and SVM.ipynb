```python
'''
Support Vector Machin SVM is a supervised learning algorithm
used for classification as well as Regression problems
used to create a hyperline or best line or decision boundary that segregate n-dimension
into classes for future entries
2 types linear and non-linear
SVM can be used for face detection,image classification and categorization etc.
'''
```

```python
from sklearn.datasets import load_iris
X,y = load_iris(return_X_y=True) # splitting input and output
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=0)
# splitting training and testing data into 70% for training and 30% for testing
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear',random_state=0)
clf.fit(xtrain,ytrain)
```

```python
ypred = clf.predict(xtest)
ypred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
from sklearn.metrics import classification_report
cr = classification_report(ytest,ypred)
print(cr)
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(ytest,ypred)
ac = ac*100
ac
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
