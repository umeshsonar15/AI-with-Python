```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df2 = df[['survived','age','fare']]
df2.fillna(1)
df2.head()
```

```python
df3 = df2.fillna(df2.mean())
```

```python
x = df3.drop("survived", axis = 1 )
y = df3["survived"]
print('shape of x', x.shape)
print('shape of y', y.shape)
```

```python
x.describe()
```

```python
x.shape
```

```python
y.shape
```

```python
ss = StandardScaler()
standard_x = ss.fit_transform(x)
standard_y = ss.fit_transform(y)
print('\nBefore Standardization x : ',x)
print('\nAfter Standardization x : ',standard_x)
print('\nBefore Standardization y : ',y)
print('\nAfter Standardization y : ',standard_y)
```

```python
mms = MinMaxScaler()
#mms = MinMaxScaler(feature_range=(0,1))
normal_x = mms.fit_transform(x)
normal_y = mms.fit_transform(y)
print('\nBefore Normalization x : ',x)
print('\nAfter Normalization x : ',normal_x)
print('\nBefore Normalization y : ',y)
print('\nAfter Normalization y : ',normal_y)
```
