```python
'''
Feature engineering is the pre-processing step of ML
which transforms raw df into features (input) to our model
Feature engineering means creating new features from existing one's
e.g
from fullname we can create firstname,mniddlename and lastname
from marks1 and marks2 we can create total marks by taking there sum.
Outlier managenent means handling the df which is an outsider compared to others
e.g
realworld example will be
0.Red ball in group of blue
1.Driving a car on footpath or on wrong side
2.A foreigner from outside of India
programming example
1.Stack Overflow
2.ArrayIndexOutOfBounds
'''
```

```python
# import pandas library and dfset
import pandas as pd
filename = '/content/drive/MyDrive/p5_feature_engineering.csv.txt'
df = pd.read_csv(filename)
df
```

```python
# creating new columns from existing columns
df['total_marks'] = df['marks1'] + df['marks2']
df
```

```python
# remove unwanted columns from dfset
df = df.drop(columns=['marks1','marks2'])
df.head(10) # to print only first 10 rows
```

```python
# students fullname from given details
df['fullname'] = df['firstname'] + df['middlename'] + df['lastname']
df
```

```python
df.info() # dfframe information columns name and there dftypes
```

```python
# Outlier management
# here rno for div A is from 1 to 64
first = 1
last = 64
print(df['rno'].isin([i for i in range(first,last)]))
```

```python
# drop the outliers from dfframe
df = df.drop(index=7) # remove outlier by index number
df = df.dropna() # remove null values
df
```

```python
'''
What is One Hot Encoding?
A one hot encoding is a representation of categorical variables as binary vectors.
This first requires that the categorical values be mapped to integer values.
Then, each integer value is represented as a binary vector that is all zero values
except the index of the integer, which is marked with a 1.
i.e.
if we have n values
a index value have assigned 1
and n-1 values have assigned 0
'''
```

```python
print(df['rno'].unique()) # check for unique value
print(df['rno'].value_counts().sum()) # total no of rno
```

```python
# One Hot Encoding
one_hot_encoded_df = pd.get_dummies(df, columns = ['rno'])
print(one_hot_encoded_df)
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc
```

```python
# Converting type of columns to category
df['rno']=df['rno'].astype('category')
df['total_marks']=df['total_marks'].astype('category')


#Assigning numerical values and storing it in another columns
df['rno_new']=df['rno'].cat.codes
df['total_marks_new']=df['total_marks'].cat.codes


#Create an instance of One-hot-encoder
enc=OneHotEncoder()

#Passing encoded columns
'''
NOTE: we have converted the enc.fit_transform()
method to array because the fit_transform method
of OneHotEncoder returns SpiPy sparse matrix
this enables us to save space when we
have huge number of categorical variables
'''
enc_df=pd.DataFrame(enc.fit_transform(df[['rno_new','total_marks_new']]).toarray())

#Merge with main dataframe df by using join
New_df=df.join(enc_df)

print(New_df)
```

```python
# log transform
import numpy as np
# Calculate natural logarithm on 'total_marks_new' column
df['natural_log'] = np.log(df['rno_new'])
df # Show the dataframe 
```

```python
# Calculate logarithm to base 2 on 'total_marks_new' column
df['logarithm_tobase2'] = np.log2(df['rno_new'])
df # Show the dataframe 
```
