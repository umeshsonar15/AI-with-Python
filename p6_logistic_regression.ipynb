```python
# importing iris dataset from sklearn.datasets
from sklearn.datasets import load_iris
features,label = load_iris(return_X_y=True)
```

```python
X = features # given input features X
X
```

```python
y = label # required output y
y
```

```python
from sklearn.linear_model import LogisticRegression as LR
classifier = LR(random_state=0) # if random_state=0 is not given output will change eac
clf = classifier
clf.fit(X,y) # training our model
```

```python
X[:2,:] # gievn input
```

```python
y[:2] # acutal output
```

```python
clf.predict(X[:2,:]) # model predicted output for given input
```

```python
clf.predict_proba(X[:2,:]) # checking probability
```

```python
y # actual output
```

```python
clf.predict(X[:,:]) # our model predicted output
# only one prediction is wrong almost accurate 97.33% accuracy
```

```python
clf.score(X,y) # checking accuracy score
```
